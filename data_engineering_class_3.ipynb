```python
customers = pd.DataFrame({'Customer Names': ['Jean Paul Paul', 'jean Jean marie Vianney', 'Cedric', 'Jeanne', 'John'],
                         'Phone Number' : ['+250722222222', '+(250)722222222', '+250-722-222-222', '0722222222',
                                          '+(250) 0722,222, 222']})

def clean_phone(phone_number):

    remove_empty = phone_number.replace(' ','')
    remove_par = remove_empty.replace('(', '').replace(')','')
    remove_dash = remove_par.replace('-','')
    remove_comma = remove_dash.replace(',','')
    
    cleaned_number = remove_comma[-10:]
    
    return cleaned_number

def clean_names(names):
    
    return ' '.join(set(names.title().split()))
```

```python
customers['cleaned_phone'] = customers['Phone Number'].apply(clean_phone)
customers['cleaned_names'] = customers['Customer Names'].apply(clean_names)
```
