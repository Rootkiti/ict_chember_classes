```python
def bonus(df):
    df['Bonus'] = df['Score']+((df['Score']*10)/100)
    return(df)
```

## 2. Aggregate

```python
df.groupby(by = ['Subject']).agg(MaxScore=('Score','max'), MinScore=('Score','min'))
```

```python
def bonus(score):
    bnus = sum(score)+sum((score)*10/100)
    return(bnus)
```
