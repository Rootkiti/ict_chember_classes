```python
def clean_area(area):
    remove_m_sqm = area.replace(' m', '').replace(' sqm', '').replace('sqm','')
    if remove_m_sqm == '':
        return None
    elif '/' in remove_m_sqm:
        wl = remove_m_sqm.split('/')
        return int(wl[0]) * int (wl[1])
    else:
        return int(remove_m_sqm)
```

```python
def clean_price(price):
    
    price = price.lower()
    if 'rwf' in price:
        cln_price = price.replace('rwf', '').replace(' rwf', '').replace('rwf ', '').replace(' ','')
        return int(cln_price)
    elif '$' in price:
        cln_price = price.replace('$', '').replace(' $', '').replace('$ ', '').replace(' ','').replace('  ','')
        return int(cln_price) * 1300
    else:
        return None
```

# INTERQUARTILE RANGE METHOD

```python
# Removung outlers
cleaned['clean_area'].plot(kind='box')
```

```python
def outliers_bounds(df,column_name):
    return lower_bound, upper_bound
#lower_bound = first quartile - 1.5 of imterquartile rangge
#upper_bound = third quartile + 1.5 of imterquartile rangge
```

```python
def outliers_bounds(df, clean_area):
    # Calculate quartiles
    first_quartile = df[clean_area].quantile(0.25)
    third_quartile = df[clean_area].quantile(0.75)
   # Calculate interquartile range
    iqr = third_quartile - first_quartile
    
    # Calculate lower and upper bounds
    lower_bound = first_quartile - 1.5 * iqr
    upper_bound = third_quartile + 1.5 * iqr
    
    return lower_bound, upper_bound
```

```python
def create_dummies(df, numerical_values, dummy_columns):
    numerical_df = df[numerical_values]
    dummies_df = pd.get_dummies(df[dummy_columns], dtype = int)
    numerical_dummies = pd.concat([numerical_df, dummies_df], axis = 1)
    return numerical_dummies
    
```
